### Cleaning Data

Menghilangkan outlier

```python
def remove_outlier(df, columns):
    for column in columns:
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

# List kolom numerik
columns_to_clean = ['instant', 'season', 'yr', 'mnth', 'holiday', 'weekday',
                    'workingday', 'weathersit', 'temp', 'atemp', 'hum',
                    'windspeed', 'casual', 'registered', 'cnt']

remove_outlier(hour_df, columns_to_clean)
remove_outlier(day_df, columns_to_clean)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana distribusi jumlah sepeda yang dipinjam pada hari kerja dan hari libur?

```python
fig = go.Figure()

fig.add_trace(go.Box(
    x=merged_df['holiday_day'].map({0: 'Hari Kerja', 1: 'Hari Libur'}),
    y=merged_df['count_day'],
    name='Jumlah Sepeda yang Dipinjam',
    boxmean=True,
    jitter=0.3,
    pointpos=-1.8,
    marker=dict(
        color='#1f77b4'
    ),
))

fig.update_layout(
    title='Distribusi jumlah sepeda yang dipinjam pada hari kerja dan hari libur',
    xaxis_title='Jenis Hari',
    yaxis_title='Jumlah Sepeda yang Dipinjam',
)

fig.show()
```

```python
# Group data berdasarkan tanggal untuk menghitung rata-rata harian
daily_avg_rentals = merged_df.groupby('weekday_day')['count_day'].mean()

hari = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']
daily_avg_rentals.index = [hari[i] for i in daily_avg_rentals.index]

fig = go.Figure()

fig.add_trace(go.Bar(
    x=daily_avg_rentals.index,
    y=daily_avg_rentals.values,
    name='Rata-rata Penyewaan Harian',
    marker=dict(
        color='#1f77b4'
    )
))

fig.update_layout(
    title='Distribusi Rata-rata Penyewaan Sepeda per Hari',
    xaxis_title='Tanggal',
    yaxis_title='Rata-rata Penyewaan',
)

fig.show()
```

### Pertanyaan 2:  Apakah cuaca berpengaruh terhadap jumlah sepeda yang dipinjam?

```python
weather_dict = {
    1: 'Cerah',
    2: 'Berawan',
    3: 'Hujan Ringan atau Salju Ringan',
    4: 'Hujan Lebat atau Salju Lebat'
}

fig = go.Figure()

for weather_code in merged_df['weather_day'].unique():
    fig.add_trace(go.Scatter(
        x=merged_df[merged_df['weather_day'] == weather_code]['temp_day'] * 41,
        y=merged_df[merged_df['weather_day'] == weather_code]['count_day'],
        mode='markers',
        name=weather_dict[weather_code],
        marker=dict(
            size=8,
            line=dict(
                width=1,
                color='#1f77b4'
            )
        )
    ))

fig.update_layout(
    title='Pengaruh cuaca terhadap jumlah sepeda yang dipinjam',
    xaxis_title='Suhu (Celsius)',
    yaxis_title='Jumlah Sepeda yang Dipinjam',
)

fig.show()
```

### Pertanyaan 3: Berapa rata-rata jumlah penyewaan jam tiap jamnya?

```python
# Group data berdasarkan hour untuk menghitung rata-rata jumlah rental
hourly_rentals = merged_df.groupby('hour')['count_hour'].mean()
trace = go.Bar(x=hourly_rentals.index,
               y=hourly_rentals.values,
               marker=dict(
                  color='#1f77b4'))
layout = go.Layout(title='Rata-rata jumlah penyewaan tiap jamnya',
                   xaxis=dict(title='Hour'),
                   yaxis=dict(title='Rata-rata Penyewaan'))

fig = go.Figure(data=[trace], layout= layout)

fig.show()
```

### Pertanyaan 4: Bagaimana distribusi jumlah sepeda yang dipinjam pada setiap bulan dan musimnya?

```python
# Group data berdasarkan month untuk menghitung total jumlah rental
casual_rentals = merged_df.groupby('month_day')['casual_day'].sum()
registered_rentals = merged_df.groupby('month_day')['registered_day'].sum()

bulan = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']
casual_rentals.index = [bulan[i-1] for i in casual_rentals.index]
registered_rentals.index = [bulan[i-1] for i in registered_rentals.index]

fig = go.Figure()

fig.add_trace(go.Bar(
    x= casual_rentals.index,
    y=casual_rentals.values,
    name='Casual',
    marker=dict(
        color='#1f77b4'
    )
))

fig.add_trace(go.Bar(
    x=registered_rentals.index,
    y=registered_rentals.values,
    name='Registered',
    marker=dict(
        color='#ff7f0e'
    )
))

fig.update_layout(
    title='Distribusi jumlah sepeda yang dipinjam pada setiap bulan',
    xaxis_title='Bulan',
    yaxis_title='Jumlah Sepeda yang Dipinjam',
    barmode='stack'
)

fig.show()
```

```python
# Group data berdasarkan musim untuk menghitung rata-rata jumlah rental
casual_rentals = merged_df.groupby('season_day')['casual_day'].mean()
registered_rentals = merged_df.groupby('season_day')['registered_day'].mean()

musim = ['Musim Semi', 'Musim Panas', 'Musim Gugur', 'Musim Dingin']
casual_rentals.index = [musim[i-1] for i in casual_rentals.index]
registered_rentals.index = [musim[i-1] for i in registered_rentals.index]

fig = go.Figure()

fig.add_trace(go.Bar(
    x= casual_rentals.index,
    y= casual_rentals.values,
    name='Casual',
    marker=dict(
        color='#1f77b4'
    )
))

fig.add_trace(go.Bar(
    x=registered_rentals.index,
    y=registered_rentals.values,
    name='Registered',
    marker=dict(
        color='#ff7f0e'
    )
))

fig.update_layout(
    title='Distribusi rata-rata sepeda yang dipinjam pada setiap musim',
    xaxis_title='Musim',
    yaxis_title='Rata-rata Sepeda yang Dipinjam',
    barmode='stack'
)

fig.show()
```

## Conclusion

- Bagaimana distribusi jumlah sepeda yang dipinjam pada hari kerja dan hari libur?
> Berdasarkan box chart, penggunaan sepeda pada rental pinjam **didominasi saat hari kerja dengan rata-rata sepeda yang dipinjam sebesar 4,557 sepeda dibandingkan dengan hari libur dengan rata-rata 3,750 sepeda yang dipinjam**. Hal ini bisa dipengaruhi oleh kepentingan atau tujuan penyewa sepeda, seperti untuk bekerja atau hanya sekedar bersepeda santai saja. Selain itu, jumlah hari kerja lebih banyak dari jumlah hari libur dalam seminggu, sehingga memungkinkan rata-rata jumlah sepeda yang dipesan lebih banyak.
Sehingga dibutuhkan visualisasi yang menunjukkan distribusi per hari dari rata-rata penyewaan sepeda yang menunjukkan bahwa **minat peminjam untuk menyewa sepeda tertinggi berada di hari Jumat dan Sabtu dengan rata-rata penyewaan sepeda sekitar sebanyak 4,703 dan 4,700.**

- Apakah cuaca berpengaruh terhadap jumlah sepeda yang dipinjam?
> Kebanyakan **penyewa sepeda memilih untuk menyewa sepeda pada hari dengan cuaca yang cerah. Selain itu, terbukti bahwa baik suhu maupun keadaan cuaca berpengaruh kepada jumlah penyewaan sepeda. Jumlah penyewaan sepeda merosot apabila suhu nya terlalu dingin dan terlalu panas, serta saat hujan ringan atau salju ringan melanda.** Apalagi saat kondisi hujan lebat, berkabut, badai atau kondisi **cuaca buruk lainnya terlihat bahwa tidak ada penyewaan sepeda sama sekali.**

- Berapa rata-rata jumlah penyewaan jam tiap jamnya?
> 3 waktu paling ideal yang menunjukkan rata-rata tertinggi penyewaan sepeda berada pada **jam 5 sore dengan rata-rata 461 sepeda dipinjam, lalu sejam setelahnya yaitu jam 6 sore dengan rata-rata sepeda dipinjam sebanyak 425, serta jam 8 pagi dengan banyak sepeda yang dipinjam sebanyak 359 sepeda.** Sehingga pemilik rental sepeda dapat menyediakan sepeda lebih banyak pada jam-jam tersebut.


- Bagaimana distribusi jumlah sepeda yang dipinjam pada setiap bulan dan musimnya?
> Berdasarkan total jumlah peminjaman sepeda, **bulan Agustus menjadi bulan dengan total terbanyak yaitu sebesar 6,673,953 peminjaman sepeda.** Yang juga bisa dilihat melalui barchart kedua bahwa hal ini berkaitan dengan rata-rata peminjaman terbanyak yang dominan berada di musim gugur. Yang berkemungkinan berkaitan dengan pengaruh suhu dan kondisi cuacanya dimana saat musim gugur suhu cenderung lebih stabil **sehingga di antara semua musim yang ada musim gugur menjadi musim dengan rata-rata penyewa sepeda terbanyak yaitu 4,450 penyewa terdaftar dan 1,204 penyewa biasa.**
